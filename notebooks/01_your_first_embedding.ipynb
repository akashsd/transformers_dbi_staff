# Your First Embedding
Embeddings turn text into numbers that preserve meaning. Similar sentences land near each other in vector space. That is what makes semantic search possible.

```python
!pip install sentence-transformers umap-learn matplotlib pandas -q
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentence = 'The officer filed a report because she was on duty.'
vector = model.encode(sentence)
print(vector.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
pair_a = model.encode(['The officer filed a report.', 'The deputy wrote up an incident.'])
pair_b = model.encode(['The officer filed a report.', 'Bananas are on the table.'])
print('similar', cosine_similarity([pair_a[0]], [pair_a[1]])[0][0])
print('unrelated', cosine_similarity([pair_b[0]], [pair_b[1]])[0][0])
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
sentences = [
    'The officer filed a report.',
    'The deputy wrote up an incident.',
    'A pothole opened on the corner.',
    'The library hosted a reading program.',
    'A fire crew handled a kitchen call.'
]
embeddings = model.encode(sentences)
coords = PCA(n_components=2).fit_transform(embeddings)
plt.figure(figsize=(6, 5))
plt.scatter(coords[:, 0], coords[:, 1])
for index, label in enumerate(sentences):
    plt.annotate(str(index + 1), (coords[index, 0], coords[index, 1]))
plt.title('Five sentences in 2D')
plt.show()
```
